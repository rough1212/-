```python
import os
import pandas as pd
OUT_CSV = "seoul_energy_use_2015_2024_from_json.csv"
CANDIDATES = [
    OUT_CSV
]
def load_first_available(paths):
    for p in paths:
        if os.path.exists(p):
            print(f"[LOAD] {p}를 불러온다.")
            return pd.read_csv(p)
    raise FileNotFoundError("대상 CSV가 현재 경로에 없다. 이전 단계의 수집 스크립트로 파일을 먼저 생성한다.")
def ensure_date(df: pd.DataFrame) -> pd.DataFrame:
    if {"year", "month"}.issubset(set(df.columns)):
        df["year"] = df["year"].astype(int)
        df["month"] = df["month"].astype(int)
        df["date"] = pd.to_datetime(
            {"year": df["year"], "month": df["month"], "day": 1}, errors="coerce"
        )
    elif "date" in df.columns:
        df["date"] = pd.to_datetime(df["date"], errors="coerce")
        df["year"] = df["date"].dt.year
        df["month"] = df["date"].dt.month
    else:
        raise ValueError("year/month 또는 date 컬럼이 없다. 최소 year와 month가 필요하다.")
    return df
def month_to_season(m: int) -> str:
    if m in (3, 4, 5):
        return "봄"
    if m in (6, 7, 8):
        return "여름"
    if m in (9, 10, 11):
        return "가을"
    return "겨울"
def add_year_and_season(df: pd.DataFrame) -> pd.DataFrame:
    df["year"] = pd.to_datetime(df["date"], errors="coerce").dt.year
    df["season"] = df["month"].apply(lambda x: month_to_season(int(x)) if pd.notna(x) else None)
    season_cat = pd.CategoricalDtype(categories=["겨울", "봄", "여름", "가을"], ordered=True)
    df["season"] = df["season"].astype(season_cat)
    return df
def main():
    df = load_first_available(CANDIDATES)
    df = ensure_date(df)
    df = add_year_and_season(df)
    print(df.head())

    print("\n=== 전처리 결과 미리보기 ===")
    cols_show = [c for c in ["date", "year", "month", "season"] if c in df.columns]
    print(df[cols_show].head(10))

    print("\n=== 결측치 점검(year/season) ===")
    print(df[["year", "season"]].isna().sum())

    print("\n=== 계절별 레코드 수 ===")
    print(df["season"].value_counts().sort_index())

    # 필요 시 저장한다.
    out_path = "seoul_energy_use_with_season.csv"
    df.to_csv(out_path, index=False, encoding="utf-8-sig")
    print(f"\n전처리된 파일을 저장했다: {out_path}")

if __name__ == "__main__":
    main()
```

```python
import os
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as Plt

# ===== 사용자 설정 =====
STUDENT_ID4 = "3505"  # 학번 뒤 4자리
CANDIDATES = [
    "seoul_energy_use_2015_2024_from_json.csv"
]
OUT_PNG = f"연도별_에너지_사용_총액_변화_{STUDENT_ID4}.png"

# ===== 유틸 =====
def normalize_columns(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.lower()
        .str.replace(r"[()\[\]\s\-/\.]", "_", regex=True)
    )
    return df

def ensure_year(df: pd.DataFrame) -> pd.DataFrame:
    if "year" not in df.columns:
        if "date" in df.columns:
            dt = pd.to_datetime(df["date"], errors="coerce")
            df["year"] = dt.dt.year
        else:
            raise ValueError("year 또는 date 컬럼이 필요합니다.")
    return df

def pick_energy_columns(df: pd.DataFrame):
    cols = list(df.columns)
    def match_any(col, patterns):
        c = col.casefold()
        return any(p in c for p in patterns)

    patt_elec  = ["elec","electric","electricity","kwh","전기"]
    patt_gas   = ["gas","가스","m3","㎥"]
    patt_water = ["water","tap","수도","상수"]
    patt_heat  = ["heat","district","heating","지역난방","난방"]

    elec = [c for c in cols if match_any(c,patt_elec)]
    gas  = [c for c in cols if match_any(c,patt_gas)]
    water= [c for c in cols if match_any(c,patt_water)]
    heat = [c for c in cols if match_any(c,patt_heat)]

    selected = sorted(set(elec + gas + water + heat))

    if not selected:
        num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
        blacklist = ["year","month","season","cnt","count","ratio","rate","avg","mean","median","min","max","std","var","총합","합계","sum"]
        selected = [c for c in num_cols if not any(b in c for b in blacklist)]

    return selected

# ===== 메인 =====
df = load_first_available(CANDIDATES)
df = normalize_columns(df)
df = ensure_year(df)

# 숫자형 변환
for c in df.columns:
    if c not in ["season"]:
        df[c] = pd.to_numeric(df[c], errors="ignore")

energy_cols = pick_energy_columns(df)

if not energy_cols:
    raise ValueError("전기/가스/수도/지역난방 또는 숫자형 컬럼을 찾지 못했습니다.")

# 총사용량 계산
df["total_energy_use"] = df[energy_cols].sum(axis=1, skipna=True)

# 연도별 집계
yearly = df.groupby("year", dropna=True, as_index=False)["total_energy_use"].sum().sort_values("year")

# 시각화
plt.figure(figsize=(9, 5))
plt.plot(yearly["year"], yearly["total_energy_use"], marker="o")
plt.title(f"연도별 에너지 사용 총액 변화 – {STUDENT_ID4}")
plt.xlabel("연도")
plt.ylabel("총 사용량")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.savefig(OUT_PNG, dpi=150)
plt.show()

print("\n[SAVE]", OUT_PNG)
print("\n[HEAD] 연도별 합계 미리보기:")
print(yearly.head())
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rcParams

GAS_COL_OVERRIDE = "gas_use_qty"
CANDIDATES = [
    "seoul_energy_use_with_season.csv",
    "seoul_energy_use_2015_2024_from_json.csv"
]
OUT_PNG = "계절별_가스_사용량_평균.png"

# ===== 한글 폰트(자동) =====
KOREAN_FONT_CANDIDATES = ["Malgun Gothic", "AppleGothic", "NanumGothic", "Nanum Gothic", "Noto Sans CJK KR"]
installed = {f.name for f in font_manager.fontManager.ttflist}
chosen_font = next((f for f in KOREAN_FONT_CANDIDATES if f in installed), "DejaVu Sans")
rcParams["font.family"] = chosen_font
rcParams["axes.unicode_minus"] = False

def load_first_available(paths):
    for p in paths:
        if os.path.exists(p):
            print(f"[LOAD] {p} 를 사용한다.")
            return pd.read_csv(p)
    raise FileNotFoundError("대상 CSV가 없다. 먼저 수집/전처리 파일을 생성한다.")

def normalize_columns(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df.columns = (
        df.columns.astype(str)
          .str.strip()
          .str.lower()
          .str.replace(r"[()\[\]\s\-/\.]", "_", regex=True)
    )
    return df

def month_to_season(m: int) -> str:
    if m in (3,4,5):   return "봄"
    if m in (6,7,8):   return "여름"
    if m in (9,10,11): return "가을"
    return "겨울"  # 12,1,2

def ensure_season(df: pd.DataFrame) -> pd.DataFrame:
    if "season" not in df.columns:
        if "month" not in df.columns:
            if "date" in df.columns:
                df["date"] = pd.to_datetime(df["date"], errors="coerce")
                df["month"] = df["date"].dt.month
            else:
                raise ValueError("season 컬럼이 없어 month 또는 date가 필요하다.")
        df["season"] = df["month"].apply(lambda x: month_to_season(int(x)) if pd.notna(x) else None)
    # 계절 순서 고정
    season_order = ["겨울", "봄", "여름", "가을"]
    df["season"] = pd.Categorical(df["season"], categories=season_order, ordered=True)
    return df

def pick_gas_column(df: pd.DataFrame) -> str:
    if GAS_COL_OVERRIDE and GAS_COL_OVERRIDE in df.columns:
        print(f"[GAS] Override 사용: {GAS_COL_OVERRIDE}")
        return GAS_COL_OVERRIDE

    # 1) 패턴 기반 1차 탐지
    gas_candidates = [
        "gas", "gas_use", "use_gas", "gasqty", "gas_qty", "gasamount", "gas_amount",
        "도시가스", "가스", "가스_사용량", "가스사용량",
    ]
    by_pattern = [c for c in df.columns if any(p in c for p in gas_candidates)]
    # 숫자형만 남긴다.
    by_pattern = [c for c in by_pattern if pd.api.types.is_numeric_dtype(df[c])]
    if by_pattern:
        print("[GAS] 패턴 일치 후보:", by_pattern)
        # 결측 적고 분산 큰 순으로 우선
        ranked = sorted(
            by_pattern,
            key=lambda c: (-df[c].notna().mean(), -df[c].std(skipna=True))
        )
        print("[GAS] 선택:", ranked[0])
        return ranked[0]

    # 2) 폴백: 숫자형 컬럼 랭킹(연/월/계절/비율/건수/평균/합계 등 제외)
    blacklist = ["year","month","season","ratio","rate","cnt","count","avg","mean","median","min","max","std","var","sum","total"]
    num_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
    num_cols = [c for c in num_cols if not any(b in c for b in blacklist)]
    # 랭킹 근거 출력
    if num_cols:
        stats = (
            pd.DataFrame({
                "col": num_cols,
                "coverage": [df[c].notna().mean() for c in num_cols],
                "std": [df[c].std(skipna=True) for c in num_cols]
            })
            .sort_values(["coverage","std"], ascending=[False, False])
        )
        print("\n[INFO] 숫자형 후보 랭킹(상위 10개):")
        print(stats.head(10).to_string(index=False))
        chosen = stats.iloc[0]["col"]
        print(f"[GAS] 폴백 선택: {chosen}")
        return chosen
    return ""

# ===== 메인 =====
df = load_first_available(CANDIDATES)
df = normalize_columns(df)
df = ensure_season(df)

print("\n[INFO] 실제 컬럼명 목록:")
print(list(df.columns))

gas_col = pick_gas_column(df)
if not gas_col:
    raise ValueError("가스 사용량 컬럼을 찾지 못했다. 위 [INFO] 목록을 보고 GAS_COL_OVERRIDE에 실제 컬럼명을 지정한다.")

# 숫자형 변환
df[gas_col] = pd.to_numeric(df[gas_col], errors="coerce")

# 계절별 평균
season_avg = (
    df.groupby("season", dropna=True)[gas_col]
      .mean()
      .reindex(["겨울","봄","여름","가을"])
)

print("\n[PREVIEW] 계절별 가스 사용량 평균:")
print(season_avg)

# 시각화(막대 + 수치 라벨)
plt.figure(figsize=(7,5))
bars = plt.bar(season_avg.index.astype(str), season_avg.values)
plt.title("계절별 가스 사용량 평균")
plt.xlabel("계절")
plt.ylabel("가스 사용량 평균")

for rect, val in zip(bars, season_avg.values):
    label = "NA" if pd.isna(val) else (f"{val:,.1f}" if abs(val) < 1000 else f"{val:,.0f}")
    plt.text(rect.get_x() + rect.get_width()/2, rect.get_height(),
             label, ha="center", va="bottom", fontsize=10)

plt.tight_layout()
plt.savefig(OUT_PNG, dpi=150)
plt.close()

print(f"\n[SAVE] {OUT_PNG}")
```
